```python
df = pd.read_csv("/kaggle/input/twitter-datacsv/twitter_data.csv")
print(df.head())
```

```python
df['labels'] = df['class'].map({0: "Hate Speech Detected", 1: "Offensive language Detected", 2: "No hate and offensive"})
print(df.head())
```

```python
import re
import string
import nltk
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer

nltk.download('stopwords')

stopword = set(stopwords.words('english'))
stemmer = PorterStemmer()

def clean(text):
    text = str(text).lower()
    text = re.sub('\[.*?\]', '', text)
    text = re.sub('https?://\S+|www\.\S+', '', text)
    text = re.sub('<.*?>', '', text)  # Remove '+' symbol from the regular expression
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
    text = re.sub('\n', '', text)
    text = re.sub('\w*\d\w*', '', text)
    text = [word for word in text.split(' ') if word not in stopword]  # Fix split() function call
    text = " ".join(text)
    text = [stemmer.stem(word) for word in text.split(' ')]
    text = " ".join(text)
    return text

df["tweet"] = df["tweet"].apply(clean)
print(df.head())
```

```python
df = df.dropna()
x = np.array(df["tweet"])
y = np.array(df["labels"])

cv = CountVectorizer()
x = cv.fit_transform(x)  # Correct method name
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42)
clf = DecisionTreeClassifier()
clf.fit(X_train, y_train)
```

```python
test_data = "This is some good stuff"
df = cv.transform([test_data]).toarray()
print(clf.predict(df))
```
